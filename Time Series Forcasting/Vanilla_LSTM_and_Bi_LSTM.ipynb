## Data Ppreparation
- inputs
[10, 20, 30, 40, 50, 60, 70, 80, 90]

- outputs
    X,	    	y
10, 20, 30		40
20, 30, 40		50
30, 40, 50		60


```python
def split_sequence(sequence,number_of_steps):
    x,y=list(),list()
    for i in range(len(sequence)):
        end_index=i+number_of_steps
        # print(end_index)
        if end_index>len(sequence)-1:
            break
        sequence_x=sequence[i:end_index]
        sequence_y=sequence[end_index]
        x.append(sequence_x)
        y.append(sequence_y)
        # print(x)
        # print(y)

    return x,y




sequence=[10,20,30,40,50,60,70,80,90]
number_of_steps =3
x,y=split_sequence(sequence,number_of_steps)
print('Input features: ',x)
print('Labels: ',y)
```

```python
for i in range(len(x)):
    print('input: ',x[i])
    print('output: ',y[i])
```

```python
x_array=np.array(x)
y_array=np.array(y)
print(x_array.shape)
print(y_array.shape)
```

```python
number_of_features=1
x_reshaped=x_array.reshape(x_array.shape[0],x_array.shape[1],number_of_features)
print(x_reshaped.shape)
x_reshaped
```

## Vanilla LSTM
A Vanilla LSTM is an LSTM model that has a single hidden layer of LSTM units, and an output layer used to make a prediction.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
model=Sequential()
model.add(LSTM(50,activation='relu',input_shape=(number_of_steps,number_of_features)))
model.add(Dense(1))
model.summary()
```

```python
model.compile(optimizer='adam',loss='mse',metrics=['accuracy'])
```

```python
history=model.fit(x_reshaped,y_array,epochs=200,verbose=1)
```

## Evaluate the model

```python
x_input=[70,80,90]
x_input=np.array(x_input)
x_input=x_input.reshape(1,number_of_steps,number_of_features)
yhat=model.predict(x_input,verbose=1)
print(yhat)
```

```python
print('Input: \n',x_input)
print('Predicted: ',yhat)
```

## Stacked LSTM
Multiple hidden LSTM layers can be stacked one on top of another in what is referred to as a Stacked LSTM model.

An LSTM layer requires a three-dimensional input and LSTMs by default will produce a two-dimensional output as an interpretation from the end of the sequence.

We can address this by having the LSTM output a value for each time step in the input data by setting the 
return_sequences=True argument on the layer. This allows us to have 3D output from hidden LSTM layer as input to the next.

```python
stacked_LSTM_model=Sequential()
stacked_LSTM_model.add(LSTM(50,activation='relu',return_sequences=True,input_shape=(number_of_steps,number_of_features)))
stacked_LSTM_model.add(LSTM(50,activation='relu'))
stacked_LSTM_model.add(Dense(1))

stacked_LSTM_model.summary()
```

```python
stacked_LSTM_model.compile(optimizer='adam',loss='mse',metrics=['accuracy'])
```

```python
history=stacked_LSTM_model.fit(x_reshaped,y_array,epochs=200,verbose=1)
```

```python
y_prediction_stacked_lstm=stacked_LSTM_model.predict(x_input,verbose=1)
y_prediction_stacked_lstm
```

```python
print('Input: \n',x_input)
print('Predicted: ',y_prediction_stacked_lstm)
```

## Bidirectional LSTM
A Bidirectional LSTM is an LSTM architecture that can improve model performance on sequence prediction problems.

On some sequence prediction problems, it can be beneficial to allow the LSTM model to learn the input sequence both forward and backwards and concatenate both interpretations.

We can implement a Bidirectional LSTM for univariate time series forecasting by wrapping the first hidden layer in a wrapper layer called Bidirectional.

```python
from tensorflow.keras.layers import Bidirectional
```

```python
bidirectional_LSTM_model=Sequential()
bidirectional_LSTM_model.add(Bidirectional(LSTM(50,activation='relu'),input_shape=(number_of_steps,number_of_features)))
bidirectional_LSTM_model.add(Dense(1))
bidirectional_LSTM_model.summary()
```

```python
bidirectional_LSTM_model.compile(optimizer='adam',loss='mse',metrics=['accuracy'])
```

```python
history=bidirectional_LSTM_model.fit(x_reshaped,y_array,epochs=200,verbose=1)
```

```python
predicted_y_bidirectional_lstm=bidirectional_LSTM_model.predict(x_input,verbose=0)
predicted_y_bidirectional_lstm
```

```python
print('Input: \n',x_input)
print('Predicted: ',predicted_y_bidirectional_lstm)
```

## Multiple layers of Bidirectional LSTMs

```python
bidirectional_LSTM_model_large=Sequential()
bidirectional_LSTM_model_large.add(Bidirectional(LSTM(50,activation='relu',return_sequences=True),input_shape=(number_of_steps,number_of_features)))
bidirectional_LSTM_model_large.add(Bidirectional(LSTM(50,activation='relu')))
bidirectional_LSTM_model_large.add(Dense(1))
bidirectional_LSTM_model_large.summary()
```

```python
bidirectional_LSTM_model_large.compile(optimizer='adam',loss='mse',metrics=['accuracy'])
```

```python
history=bidirectional_LSTM_model_large.fit(x_reshaped,y_array,epochs=200,verbose=1)
```

```python
predicted_y_bidirectional_lstm_large=bidirectional_LSTM_model_large.predict(x_input,verbose=0)
predicted_y_bidirectional_lstm_large
```

```python
print('Input: \n',x_input)
print('Predicted: ',predicted_y_bidirectional_lstm_large)
```
